제 전공은 컬처앤테크놀로지융합전공이며, 콘텐츠 전문가 양성 학과이기 때문에 콘텐츠 시장의 정보가 굉장히 관련이 많겠다고 판단해
한국콘텐츠진흥원의 국가별 콘텐츠 시장 정보 공공데이터를 활용한 검색 프로그램을 구현했습니다.

안내 사항에 전공 관련 데이터를 찾기 힘들 때 다른 데이터와 함께 마크다운셀로 설명 하라 하셨지만
저는 이 데이터가 왜 전공과 관련이 있는지 스스로 생각해보고 싶었고 그 고민한 결과를 설명 드리고 싶었습니다.
그리고 해당 프로그램이 초급자인 제 실력으로 작성되었기에 전체적으로 어떤 프로그램인지, 대략 어떤 방식으로 구현되었는지 설명 드리고 싶습니다.

제 전공인 콘텐츠 분야에서는 이러한 정보 검색 시 주로 "어떤 장르"가 어느 정도의 "성장율과 수익성"을 가지고 있는 지가 중요 지점이기 때문에 
검색한 국가별 정보 내에서 성장율과 상승세, 하락세였던 장르들을 따로 필터링해 리스트에 저장한 후 출력해서 가시화 하고,
추가적으로 해당 국가 콘텐츠 분야의 전반적인 평균 성장율을 계산해 출력하도록 구현했습니다.

전반적으로 실제 프로그램과 같은 느낌을 주기 위해 입력문과 입력에 따른 반복문 동작을 활용했습니다.

```python
# 실제 프로그램과 유사하게 인삿말과 실수로 프로그램을 켰을 수도 있다는 가정하에 프로그램을 정말 실행할지 질문
print("국가별 콘텐츠 시장 정보 검색 서비스에 오신 것을 환영합니다!\n")
input_want_ = input("국가별 콘텐츠 시장 정보 검색 서비스를 이용 하시겠습니까? [Y/y] : ")

if input_want_ == 'Y' or input_want_ == 'y':
    #csv 파일을 불러와서 한 줄씩 저장하는 부분은 반복문에 의해 반복될 시 메모리가 낭비될 수 있기 때문에 반복문 바깥에서 실행
    import csv

    data_file = '국가별콘텐츠시장통계정보.csv'
    data_pointer = open(data_file, 'r', encoding='utf-8')

    data_list = []
    data_reader = csv.reader(data_pointer)
    for one_line in data_reader:
        data_list.append(one_line)

    #입력문의 대답에 따라 반복문을 실행 후 리스트를 초기화하고 정보를 다시 입력 받음
    while True:
        growth_data_positive = []
        growth_data_negative = []
        positive_genre = []
        negative_genre = []

        col = data_list[0]
        growth_index = col.index("성장율 데이터")
        genre_index = col.index("장르이름")

        # 특정 나라의 콘텐츠 시장 콘텐츠 정보
        input_country = input("\n콘텐츠 시장 정보를 찾아보고 싶은 국가명 : ")
        for one_line in data_list[1:]:
            if input_country in one_line[col.index('\ufeff나라이름')]:
                for index, country in enumerate(col):
                    if index == growth_index and one_line[index] != '': #데이터 중에서 
                        if float(one_line[growth_index]) > 0:
                            growth_data_positive.append(one_line[growth_index])
                            positive_genre.append(one_line[genre_index])
                        else:
                            growth_data_negative.append(one_line[growth_index])
                            negative_genre.append(one_line[genre_index])

                    print(f"{country} : {one_line[index]}")

                print("*" * 20)
                print("~" * 20)

        # 상승세인 장르들을 중복 없이 종류별로 하나씩만 담는 리스트
        only_positive_genre = []
        for genre in positive_genre:
            if genre not in only_positive_genre:
                only_positive_genre.append(genre)

        positive_genre = only_positive_genre

        # 하락세인 장르들을 중복 없이 종류별로 하나씩만 담는 리스트
        only_negative_genre = []
        for genre in negative_genre:
            if genre not in only_negative_genre:
                only_negative_genre.append(genre)

        negative_genre = only_negative_genre
        
        # 각각 상승세, 하락세였던 해당 국가의 장르 목록을 중복없이 나열하고, 성장율 데이터의 평균을 구할 때 사용될 데이터 목록 가시화
        #->잘못된 데이터 추출 식별가능
        print(f"상승세였던 장르 목록 : {positive_genre}")
        print(f"하락세였던 장르 목록 : {negative_genre}")
        print(f"상승세였던 성장율 데이터 목록 : {growth_data_positive}")
        print(f"하락세였던 성장율 데이터 목록 : {growth_data_negative}")
        
        #해당 국가의 카테고리 기준으로 모두 따로 계산된 성장율 데이터들의 평균을 계산하기 위해 새로운 리스트를 만들어 상승세와 하락세 리스트를 합침 
        unique_growth_data = growth_data_positive + growth_data_negative
        data_sum = 0
        average = 0
        count = 0
        for i in range(0, len(unique_growth_data)):
            data_sum += float(unique_growth_data[i])
            count += 1
        average = data_sum / count
        print(f"{input_country}의 성장율 데이터 평균 : {average}")

        # 원하면 다른 국가들의 콘텐츠 시장 성장율도 반복을 통해 확인 가능
        input_want_again = input("다른 국가의 콘텐츠 시장 정보를 찾아보시겠습니까? [Y/y] : ")

        if input_want_again == 'Y' or input_want_again == 'y':
            continue
        else:
            print("콘텐츠 시장 정보 찾기 서비스를 종료합니다.")
            break

#실수로 프로그램을 구동시켰을 시에 다른 대답을 입력해 프로그램 종료
else:
    print("콘텐츠 시장 정보 찾기 서비스를 종료합니다.")
```
